```python
# Dependencies

from sqlalchemy import create_engine
import pandas as pd
import numpy as np
```

```python
# Load files
nchs_data = "nchs.csv"
awram_data = "awram.csv"

# Read both datasets and store into Pandas Data Frames
nchs_df = pd.read_csv(nchs_data)
awram_df = pd.read_csv(awram_data)
```

```python
nchs_df.head()
```

```python
# Check name of columns

nchs_df.columns
```

```python
# Import just the columns required. Age, Injury, Year, Death, Population, Sex, Year

firearm1 = nchs_df[["Age group (years)", "Injury intent", "Injury mechanism","Deaths",
                            "Population", "Sex", "Race",
                            "Year"]]
firearm1.head()
```

```python
# Filter dataset for just 2015 

df_2015 = firearm1.loc[firearm1["Year"] == 2015]
df_2015.head()
```

```python
# Filter dataset for just Firearms

df_gun = df_2015.loc[df_2015["Injury mechanism"] == 'Firearm']
df_gun.head()
```

```python
df_sex = df_gun[(df_gun["Sex"] == 'Male') | (df_gun["Sex"] == 'Female')]
df_sex.head()
```

```python
# Now, filter by race and choose "All Races". This removes duplicates

df_race = df_sex.loc[df_sex["Race"] == 'All races']
df_race.head()
```

```python
# Main table. Rename because (years) is mistaken for a function

renamed_table = df_race.rename(columns={"Age group (years)": "Age_group", "Injury intent": "Injury_type",
                                        "Injury mechanism": "Outcome",
                                        "Deaths": "Count"})
renamed_table.head()
```

```python
# Step 1 of 2 to create a new table for poulation
# Filter for just "All Intentions"

df_pop1 = renamed_table.loc[df_race["Injury intent"] == 'All Intentions']
df_pop1.head()
```

```python
# Carved out Population groupings as a separate table
# It needed age grouping and sex as lookup columns
# Re-indexed
# Problem with "index" being imported

df_pop2 = df_pop1[["Age_group", "Sex", "Population"]]
df_pop2.reset_index(inplace=True)
df_pop2.head()
```

```python
# Back to main Table. Filter by "Injury intent" to delete the total "All Intention"
# This removes duplicates

df_intention = renamed_table.loc[(df_race["Injury intent"] == 'Unintentional') |  
                           (df_race["Injury intent"] == 'Suicide') | 
                          (df_race["Injury intent"] == 'Homicide') | 
                           (df_race["Injury intent"] == 'Undetermined') | 
                           (df_race["Injury intent"] == 'Legal intervention/war')
                          ]

df_intention.head()
```

```python
#Get list of values in a column so I can use same Age Group bins for second database

males = df_intention.loc[(df_intention["Injury_type"] == 'Unintentional') & (df_intention["Sex"] == 'Male')]
males.head()
males.Age_group.unique
```

```python
# Filter by age categories to delete the total "All Ages". This removes duplicates

df_all_ages = df_intention.loc[(df_intention["Age_group"] == '< 15') |  
                           (df_intention["Age_group"] == '15-24') | 
                          (df_intention["Age_group"] == '25-64') | 
                           (df_intention["Age_group"] == '45-64') | 
                           (df_intention["Age_group"] == '65-74') | 
                           (df_intention["Age_group"] == '75+')
                          ]

df_all_ages.head()
```

```python
# Drop 3 columns that are unnecessary, (previously filtered for), now they can be dropped
# Delete Population / Race / Year
# I have kept Firearms as column to merge with Product from 2nd database

df_all = df_all_ages[["Age_group", "Injury_type", "Outcome", "Count", "Sex"]]
df_all.head()
```

```python
# Bin ages to match the data from NCHS

bins = [0, 14, 24, 44, 64, 74, 120]
age_labels = ["< 15", "15–24", "25–44", "45–64", "65–74", "75+"]


accident["Age_group"] = pd.cut(accident["age"], bins, labels=age_labels)
accident.head()
```

```python
# Rename table

renamed_accident = accident.rename(columns={"sex": "Sex", "diagnosis": "Outcome",
                                        "product": "Injury_type"})
renamed_accident.head()
```
